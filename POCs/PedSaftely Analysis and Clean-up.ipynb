```python
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt

import pandas as pd
import seaborn as sns
```

```python
# - load the dataset: ../data/international-airline-passengers.csv
df = pd.read_csv('C:/Users/Preeti/Github/CapstoneProject/data/PedSaftey.csv')
```

```python
df.columns
```

```python
df.drop(['Additional_COMMENTS','STRSEGID.1'], axis =1, inplace=True)
```

```python
df.columns
```

```python
df.drop_duplicates(keep='first', inplace=True)
```

```python
df.reset_index( inplace=True)
```

```python
df
```

```python
df_addC = df[['REQUESTID','STRSEGID','Additional_COMMENTS']]
df_addC.fillna('', inplace=True)
```

```python
df2 = df_addC.groupby('REQUESTID').apply(lambda x: ' '.join(x.Additional_COMMENTS))
```

```python
df3 = df2.to_frame(name=None)
```

```python
df3
```

```python
df['REQUESTID'] = df['REQUESTID'].astype(int)
```

```python
df3 = df3.reset_index()
df3['REQUESTID'] = df3['REQUESTID'].astype(int)
```

```python
#result = pd.concat([df, df2], axis=1, keys=['REQUESTID'])
result = pd.merge(df, df3, on='REQUESTID')
result
```

```python
result.columns
```

```python
result.rename(columns={'Request type cleaned':'Request_type'}, inplace=True)
```

```python
result
```

```python
result.to_csv('../data/PedSaftelyCleaned.csv', index=False)
```

```python
requestType = result.groupby(['Request_type']).count()
requestType.reset_index()
requestType['Request_type'] = requestType.index
```

```python
sns.barplot(y="Request_type", x="REQUESTID", data=requestType,palette="Set1")
```

```python
requestType = result.groupby(['SNA_NAME']).count()
requestType.reset_index()
requestType['SNA_NAME'] = requestType.index
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.barplot(y="SNA_NAME", x="REQUESTID", data=requestType,palette="Set1")
```
